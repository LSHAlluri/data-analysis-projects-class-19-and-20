```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv",on_bad_lines='skip')
# df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
# df.columns


```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.

# Use strip() Method: If there are any leading or trailing spaces in the column names, you can remove them:
df.columns = df.columns.str.strip()


scatter_plot = pd.DataFrame({
    "avg_rating": df['average_rating'],
    "pages": df['num_pages']
})

sns.scatterplot(data=scatter_plot, x="avg_rating", y="pages")
plt.xlabel("Average Rating")
plt.ylabel("Number of Pages")
plt.title("Scatter Plot of Average Rating vs Number of Pages")
plt.show()
```

```python
# Use your Python skills to create a visualization here.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='average_rating', data = df, marker='o')
plt.title('Year vs Average rating')
plt.xlabel('Years')
plt.ylabel('Average rating')
plt.grid(True)
plt.show()

```
